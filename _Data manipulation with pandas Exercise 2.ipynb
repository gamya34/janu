# Chipotle data set from github

```python
url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

```python
# Import the necessary libraries
import pandas as pd
```

```python
#Import the dataset from the given url
#Assign it to a variable called chipo.
chipo=pd.read_csv(url,sep='\t')
```

```python
chipo.head(2)
```

```python
chipo['item_price']=chipo['item_price'].str.replace("$","")
chipo['item_price']=pd.to_numeric(chipo['item_price'])
```

```python
#What is the price of each item?
#print a data frame with only two columns item_name and item_price
chipo.groupby('item_name')['item_name','item_price'].mean().head(2)
```

```python
#Sort by the name of the item
chipo.sort_values("item_name")
```

```python
#What was the quantity of the most expensive item ordered?
chipo.sort_values('item_price').max()
#chipo.sort_values("item_price", ascending = False).head(1)
```

```python
#How many times were a Veggie Salad Bowl ordered?
#chipo.query('item_name=="Veggie Salad Bowl"').shape[0]
len(chipo.query('item_name=="Veggie Salad Bowl"'))
```

```python
#How many times people orderd more than one Canned Soda?
chipo.query('item_name=="Canned Soda" & quantity>1').shape[0]
```
